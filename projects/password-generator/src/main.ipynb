```python
import random
import string

class RandomGenerator(PasswordGenerator):
    def __init__(self):
        super().__init__()
        self.numbers = range(10)
        self.letters = string.ascii_letters
        self.symbols = string.punctuation
    
    def generate(self, chars=16, num=False, symbol=False):
        if num == False and symbol == False:
            for char in range(chars):
                self.password += random.choice(self.letters) 
        elif num == True and symbol == False:
            char_num = random.choice(range(1, chars))
            num_num = chars - char_num 
            for char in range(char_num):
                self.password += random.choice(self.letters) 
            for num in range(num_num):
                self.password += str(random.choice(self.numbers))
                
            self.password = ''.join(random.sample(self.password, len(self.password)))
            
        elif  num == False and symbol == True:
            char_num = random.choice(range(1, chars))
            symbol_num = chars - char_num 
            for char in range(char_num):
                self.password += random.choice(self.letters) 
            for symbol in range(symbol_num):
                self.password += random.choice(self.symbols)
                
            self.password = ''.join(random.sample(self.password, len(self.password)))
            
        else:
            char_num = random.choice(range(1, chars-1))
            num_num = random.choice(range(1, chars-char_num)) 
            symbol_num = chars - num_num - char_num
            for char in range(char_num):
                self.password += random.choice(self.letters) 
            for num in range(num_num):
                self.password += str(random.choice(self.numbers))
            for symbol in range(symbol_num):
                self.password += random.choice(self.symbols)
                
            self.password = ''.join(random.sample(self.password, len(self.password)))
        
        return self.password            
```
