```python
class Contact:
    def __init__(
        self,
        first_name: str | None = None,
        last_name: str | None = None,
        phone: str | None = None,
        email: str | None = None,
        tags: list[str] | None = None,
        notes: str | None = None,
    ):
        self.first_name = first_name
        self.last_name = last_name
        self.phone = phone
        self.email = email
        self.tags = tags or []
        self.notes = notes
```

```python
mehdi = Contact()
print(mehdi)
```
